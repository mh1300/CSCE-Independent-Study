```python
# Loading Dataset
datasetsDir = '../datasets/'

vhdrPath = '../datasets/Lemon/sub-010002/RSEEG/sub-010002.vhdr'

print(vhdrPath)
raw = mne.io.read_raw_brainvision(vhdr_fname=vhdrPath)
```

```python
# Add montage to each raw file - https://www.nature.com/articles/sdata2018308
montage = mne.channels.make_standard_montage('standard_1005') 

# Read in all files
raws = {}

lemonDir = Path('../datasets/Lemon')
for subject in lemonDir.iterdir():
    if 'sub' in str(subject):
        subID = subject.name
        subjectEEGDir = subject / 'RSEEG' / f'{subID}.vhdr'

        raw = mne.io.read_raw_brainvision(vhdr_fname=str(subjectEEGDir))
        # raw.set_channel_types({'VEOG': 'eog'}) # Set the one non-eeg channel to eog
        raw.drop_channels(['VEOG'])
        raw.set_montage(montage)
        raws[subID] = raw
```

```python
print(raws[0].get_data().max())
print(raws[0].get_data().min())
# raws[0].plot(n_channels=5)
```

```python
# Meta data for bids path
task = 'RestEOEC'
bids_root = Path('../datasets/Bids_LEMON')

# Remove the directory before creating a new one
if op.exists(bids_root):
    shutil.rmtree(bids_root)
```

```python
# For all existing files, save the bids
for id, raw in raws.items():
    bids_path = BIDSPath(subject=id[4:], task=task, root=bids_root)
    write_raw_bids(raw, bids_path, overwrite=True)
```
